```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
POSITIONS = ["PG", "SG", "PF", "SF", "C"]
DATAFRAMES = {} # Array containing all datas by positions
```

```python
# Get all datas by positions
for pos in POSITIONS:
    DATAFRAMES[pos] = pd.read_excel(f"../DATAS/{pos}_DATAS.xlsx", index_col=0) 
```

```python
DATAFRAMES["PG"].columns
```

```python
"""
    We defined 7 important categories :
        1 - Offensive Versatile: Has high values in several offensive stats, such as PPG, PPS, APG, and high ORtg.
        2 - Defensive Versatile: Has high values in several defensive stats, such as SPG, BPG, DRB%, and low DRtg.
        3 - Play maker: Characterized by high APG, high AST%, and low TOV%.
        4 - Dominant Rebounder: Focuses on high RPG, high TRB%, and possibly significant BPG.
        5 - Efficient Player: Characterized by high TS%, high eFG%, and possibly high PPS.
        6 - Game Controller: Highlights a high AST%, low TOV%, and perhaps a moderate USG%.
        7 - Balanced Player: Mean of all the precedent categories.
    For each of those categories and each positions, we will cluster our players and rate them through TIER
"""
```

```python
# Plotting test for Offensive Versatile
ov_x = DATAFRAMES["PG"]["PPG"]
ov_y = DATAFRAMES["PG"]["PPS"]
ov_z = DATAFRAMES["PG"]["APG"]
ov_c = DATAFRAMES["PG"]["ORtg"]
```

```python
ov_fig = plt.figure(figsize=(10, 8))
ov_ax = ov_fig.add_subplot(111, projection='3d')
ov_img = ov_ax.scatter(ov_x, ov_y, ov_z, c=ov_c, cmap=plt.hot())
for player_name, xi, yi, zi in zip(DATAFRAMES["PG"].index, ov_x, ov_y, ov_z):
    ov_ax.text(xi, yi, zi, player_name)
    
ov_fig.colorbar(ov_img)
plt.show()
```

```python
# Plotting test for Defensive Versatile
dv_x = DATAFRAMES["PG"]["SPG"]
dv_y = DATAFRAMES["PG"]["BPG"]
dv_z = DATAFRAMES["PG"]["DRB%"]
dv_c = DATAFRAMES["PG"]["DRtg"]
```

```python
dv_fig = plt.figure(figsize=(10, 8))
dv_ax = dv_fig.add_subplot(111, projection='3d')
dv_img = dv_ax.scatter(dv_x, dv_y, dv_z, c=dv_c, cmap=plt.hot())
for player_name, xi, yi, zi in zip(DATAFRAMES["PG"].index, dv_x, dv_y, dv_z):
    dv_ax.text(xi, yi, zi, player_name)
    
dv_fig.colorbar(dv_img)
plt.show()
```
