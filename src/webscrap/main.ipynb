```python
import os
import time
import pandas as pd
```

```python
POSITIONS = ["PG", "SG", "SF", "PF", "C"]
```

```python
def get_basic_stats():
    stats = {}
    for pos in POSITIONS:
        url = f"https://basketball.realgm.com/nba/stats/Historical_Active/Averages/Qualified/player/{pos}/asc/1/Regular_Season"
        tables = pd.read_html(url)
        df = tables[-1]
        del df["#"]
        del df["Team"]
        del df["GP"]
        stats[pos] = df
    return stats
```

```python
basic_stats_by_pos = get_basic_stats()
```

```python
basic_stats_by_pos['PG']
```

```python
def get_advanced_stats():
    stats = {}
    for pos in POSITIONS:
        url = f"https://basketball.realgm.com/nba/stats/Historical_Active/Advanced_Stats/Qualified/player/{pos}/asc/1/Regular_Season"
        tables = pd.read_html(url)
        df = tables[-1]
        del df["#"]
        del df["Team"]
        del df["Player"]
        stats[pos] = df
    return stats
```

```python
adv_stats_by_pos = get_advanced_stats()
```

```python
adv_stats_by_pos["PG"]
```

```python
ALL_STATS = {}
for pos in POSITIONS:
    df = pd.concat([ basic_stats_by_pos[pos], adv_stats_by_pos[pos] ], axis=1)
    ALL_STATS[pos] = df
```

```python
if(not os.path.exists(f"../DATAS")):
    os.makedirs(f"../DATAS")
```

```python
for pos in POSITIONS:
    if os.path.exists(f"../DATAS/{pos}_DATAS.xlsx"):
        continue
    ALL_STATS[pos].to_excel(f'../DATAS/{pos}_DATAS.xlsx', index=False)
```
